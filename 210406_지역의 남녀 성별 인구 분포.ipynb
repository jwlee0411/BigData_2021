```python
import matplotlib.pyplot as plt
import csv

plt.bar([0,10,3,8,74], [1,2,3,8,11])
plt.show()

f = open('/content/drive/MyDrive/data/age.csv', 'r', encoding='cp949')
data = csv.reader(f)
result = []

for row in data:
  if '도곡2동' in row[0]:
    for i in row[3:104]:
      result.append(int(i))


plt.figure(figsize = (10,5))
plt.bar(range(101), result)
plt.show()
```

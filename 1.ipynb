```
import csv
max_temp = -999
max_date = ''
f = open('/content/drive/MyDrive/data/seoul.csv', 'r', encoding='cp949' )
data = csv.reader(f, delimiter = ",")
```

```
print(data)
```

```
for row in data:
  print(row)
```

```
header = next(data)

for row in data:
  if row[-1] == '':
    row[-1] = -999
  row[-1] = float(row[-1])
  if max_temp < row[-1]:
    max_temp = row[-1]
    max_date = row[0]

print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은', max_date, ', ', max_temp)
```

```python
import csv
f = open('/content/drive/MyDrive/subway_fee.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
next(data)
label = ['유임승차', '유임하차', '무임승차', '무임하차']
max = [0] * 4
max_station = [''] * 4
for row in data:
  for i in range(4, 8):
    row[i] = int(row[i])
  
    if row[i] > max[i-4] :
      max[i-4] = row[i]
      max_station[i-4] = row[3] + ' ' + row[1]

    # print(row)

for i in range(4):
  print(label[i] + ':' + max_station[i], max[i])
```

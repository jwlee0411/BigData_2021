```python
import csv
import matplotlib.pyplot as plt


f = open('/content/drive/MyDrive/data/seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)
result = []
result2 = []

for row in data:
  if row[-1] != '':
    date = row[0].split('-')
    if int(date[0]) >= 2003:
      if date[1] == '04' and date[2] == '11':
        result.append(float(row[-1]))
        result2.append(float(row[-2]))


plt.title("birth_ondo")
plt.plot(result, color = 'wheat', label = 'high')

plt.plot(result2, color = 'hotpink', label = 'low')
plt.legend(loc=1)
plt.show()
```

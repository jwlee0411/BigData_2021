```python
!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf

import matplotlib.pyplot as plt
import csv
plt.rc('font', family='NanumBarunGothic')  # 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지





f = open('/content/drive/MyDrive/data/gender.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

male = []
female = []

k = input()



for row in data:
  if k in row[0]:
    for i in row[3:104]:
      male.append(int(i.replace(',', '')))
    for i in row[106:]:
      female.append(int(i.replace(',', '')))
    break
    

sMale = sum(male)
sFemale = sum(female)

bloodtype = [sMale, sFemale]
label = ['Nam', 'Yeo']

plt.figure(figsize = (10, 10))
plt.title('Nam_Yeo_Seongbyeul_Ingu_Bunpo')

plt.pie(bloodtype, labels = label, autopct = '%.1f%%')

plt.show()
```

```python
import numpy as np

col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print(array1.shape);
```

```python
# list -> dataframe
df_list1 = pd.DataFrame(array1, columns = col_name1)
print(df_list1)
```

```python
# ndarray -> dataframe
df_array1 = pd.DataFrame(array1, columns = col_name1)
print(df_array1)
```
